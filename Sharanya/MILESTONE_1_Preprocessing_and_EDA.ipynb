```python
most_common_reason = df['DelayReason'].mode()[0]
df['DelayReason'] = df['DelayReason'].fillna(most_common_reason)
```

```python
df.isnull().sum()
```

```python
df['ScheduledDeparture'] = pd.to_datetime(df['ScheduledDeparture'])
df['ActualDeparture'] = pd.to_datetime(df['ActualDeparture'])
df['ScheduledArrival'] = pd.to_datetime(df['ScheduledArrival'])
df['ActualArrival'] = pd.to_datetime(df['ActualArrival'])
```

```python
df.info()
```

```python
# Extract month, day, and hour
df['Month'] = df['ScheduledDeparture'].dt.month
df['DayOfWeek'] = df['ScheduledDeparture'].dt.day_name()
df['Hour'] = df['ScheduledDeparture'].dt.hour
df['ScheduledDeparture'] = pd.to_datetime(df['ScheduledDeparture'], format='%H:%M').dt.time
```

```python
df['Route'] = df['Origin'] + '-' + df['Destination']
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f'Top categories in {col}')
    plt.ylabel('Count')
    plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Airline', y='DelayMinutes', data=df)
plt.title('Airline vs Departure Delay')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='DayOfWeek', y='DelayMinutes', data=df)
plt.title('Day of Week vs Arrival Delay')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Distance', y='DelayMinutes', data=df, alpha=0.3)
plt.title('Distance vs Arrival Delay')
plt.show()
```

```python
# Boxplot for detecting outliers in delay columns
plt.figure(figsize=(6,4))
sns.boxplot(x=df['DelayMinutes'])
plt.title('Outlier Detection - Departure Delay')
plt.show()

```

```python
Q1 = df['DelayMinutes'].quantile(0.25)
Q3 = df['DelayMinutes'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df['DelayMinutes'] < lower_limit) | (df['DelayMinutes'] > upper_limit)]
print("Number of outliers:", len(outliers))
print(outliers[['Airline', 'DelayMinutes', 'DelayReason']].head())
```
